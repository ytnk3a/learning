## .4 確率的試行のシミュレーション

```julia
using Distributions

bern = Bernoulli(0.5)
n = rand(bern, 10)
```

```julia
rand(Bernoulli(0.3), 10) 
```

```julia
bag(x::Bool) = x == 1 ? "A" : "B"
ball(x::Bool) = x == 1 ? "赤" : "白"

x = bag.(rand(Bernoulli(0.3), 10) )
```

```julia
function sample()
    x = bag(rand(Bernoulli(1//2)))

    μ = x == "A" ? 1//5 : 3//5

    y = ball(rand(Bernoulli(μ)))
    
    x, μ, y
end

for _ in 1:10
    x, μ, y = sample()
    println("袋: $(x), 玉:$(y)")
end
```

### .4.2 周辺確率の計算

```julia
maxiter = 1000000
result = []

for _ in 1:maxiter
    x, μ, y = sample()
    push!(result, y)
end

mean(result .== "赤")
```

### .4.3 条件付き確率の計算

```julia
y_obs = "赤"
maxiter = 1000000
x_results = []

for _ in 1:maxiter
    x, μ, y = sample()

    y == y_obs && push!(x_results, x)

end

println("acceptance rate = $(length(x_results) / maxiter)")
println("p(x=A|y=赤) : approx = $(mean(x_results .== "A")) (exact = $(1/4))")
println("p(x=B|y=赤) : approx = $(mean(x_results .== "B")) (exact = $(3/4))")
```

### .4.4 複数のデータがある場合

```julia
function sample(N)
    x = bag(rand(Bernoulli(1//2)))

    μ = x == "A" ? 1//5 : 3//5
    y = ball.(rand(Bernoulli(μ), N))
    
    x, μ, y
end
```

```julia
maxiter = 1000000
y_obs = ["赤", "赤", "白"]
x_results = []

for _ in 1:maxiter
    
    x, μ, y = sample(3)
    y == y_obs && push!(x_results, x)

end

println("acceptance rate = $(length(x_results) / maxiter)")
println("p(x=A|y1=赤, y2=赤, y3=白) : approx = $(mean(x_results .== "A")) (exact = $(2/11))")
println("p(x=B|y1=赤, y2=赤, y3=白) : approx = $(mean(x_results .== "B")) (exact = $(9/11))")
```

```julia
maxiter = 1000000
x_results = []

for _ in 1:maxiter
    x, μ, y = sample(3)

    y == y_obs && push!(x_results, x)

    # sum(y .== "赤") == sum(y_obs .== "赤") && push!(x_results, x)
    sum(y .== "赤") == 2 && push!(x_results, x)

end

println("acceptance rate = $(length(x_results) / maxiter)")
println("p(x=A|y1=赤, y2=赤, y3=白) : approx = $(mean(x_results .== "A")) (exact = $(2/11))")
println("p(x=B|y1=赤, y2=赤, y3=白) : approx = $(mean(x_results .== "B")) (exact = $(9/11))")
```
